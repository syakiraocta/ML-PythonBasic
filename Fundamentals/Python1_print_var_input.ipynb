# Operator

### Operator Aritmetika

| Operator | Arti                                                                                          | Contoh                  |
|----------|--------------------------------------------------------------------------------------------------|--------------------------|
| +        | Penambahan                                                                   | x + y                 |
| -        | Pengurangan                                              | x - y                  |
| *        | Perkalian                                                                            | x * y                    |
| /        | Pembagian                                 | x / y                    |
| %        | Modulo - Sisa Pembagian                                 | x % y (sisa pembagian of x/y) |
| //       | Pembagian dan dibulatkan ke bawah | x // y                   |
| **       | Eksponen - Pangkat                                             | x**y (x pangkat y)  |

````python
print(2 ** 3)
print(2 ** 3.)
print(2. ** 3.)
print(2. ** 3)

#(**) berarti pangkat
# tanda(.) artinya float/desimal
````

````python
print(10 / 3)
print(10 // 3)

print(6 / 3.)
print(6 // 3.)

print(-4 + 4)
print(-4. + 8)

#(/) Pembagian biasa jadi hasilnya bisa float (decimal)
#(//) Pembagian yang diambil angka bulatnya aja (interger)
````

### Urutan Operator

| Prioritas| Operator |        |
|----------|----------|--------|

| 2        | **       |        |
| 3        | *,/,//,% |        |
| 4        | +, -     | binary |

````python
2 + 3 *5
````

````python
print(9 % 6 % 2)
print(9 % 6)
print(3 % 2)

print(2 ** 2 ** 3)
print(2 ** 2)
print(2 ** 8)
````
