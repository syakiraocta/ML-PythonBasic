````python
def bmi(weight, height):
    return weight / height ** 2

print(bmi(65, 1.69))
````

Basic perhitungan BMI

## Faktorial

````python
def factorialFun(n):
    if n < 0: # Tidak ada faktorial untuk angka negatif
        return None
    if n < 2: #Faktorial dibawah 2 hanya menghasilkan angka 1
        return 1

    product = 1
     #variabel untuk menampung hasil perkalian, bernilai 1 sebagai nilai awal
    for i in range(2, n + 1): #start menghitung dari angka 2 dan stop di n
        product *= i
        #Progam mengalikan nilai yang ada didalam product dengan nilai i (urutan berikutnya)
    return product

#Call the Function here
for n in range(1, 15): # testing
    print(n, factorialFun(n))
````

## Fibonacci

````python
def fib(n):
    if n < 1: #tidak ada urutan dari minus
         return None
    if n < 3: #urutan 1 dan 2 dalam fibonaci pasti angka 1
        return 1

    elem1 = elem2 = 1 #urutan pertama dan kedua sudah pasti bernilai 1
    sum = 0
    for i in range(3, n + 1): #menghitung mulai dari urutan ke-3hingga n
        sum = elem1 + elem2  # operasi untuk menghasilkan angka baru
        elem1, elem2 = elem2, sum #menggeser posisi urutan dari urutan 1 - n
    return sum

#Call the Function here
for n in range(1, 10): #testing
    print(n, ">", fib(n))
````

## Recursion

Fungsi yang memanggil dirinya sendiri

````python
#recursion 1
def fib(n):
    if n < 1:
        return None
    if n < 3:
        return 1
    return fib(n - 1) + fib(n - 2)
  #Untuk menghasilkan output dengan urutan tertentu, maka harus melakukan penjumlahan 2 urutan sebelumnya

for n in range(1, 10): # testing
    print(n, "->", fib(n))

#recursion 2

def fun(a):
    if a > 30:
        return 3
    else:
        return a + fun(a + 3)


print(fun(25))
````

## Dictionaries

In Python's world, the word you look for is named a key. The word you get from the dictionary is called a value.

Dictionaries are unordered*, changeable (mutable), and indexed collections of data. (*In Python 3.6x dictionaries have become ordered by default.

This means that a dictionary is a set of key-value pairs. Note:

- each key must be unique - it's not possible to have more than one key of the same value;
- a key may be data of any type: it may be a number (integer or float), or even a string;
- a dictionary is not a list - a list contains a set of numbered values, while a dictionary holds pairs of values;
- the len() function works for dictionaries, too - it returns the numbers of key-value elements in the dictionary;
- a dictionary is a one-way tool - if you have an English-French dictionary, you can look for French equivalents of English terms, but not vice versa.

### Tuples and Dictionaries work Together



- line 1: create an empty dictionary for the input data; the student's name is used as a key, while all the associated scores are stored in a tuple (the tuple may be a dictionary value - that's not a problem at all)
- line 3: enter an "infinite" loop (don't worry, it'll break at the right moment)
- line 4: read the student's name here;
- line 5-6: if the name is exit, leave the loop;
- line 8: ask for one of the student's scores (an integer from the range 1-10)
- line 10-11: if the student's name is already in the dictionary, lengthen the associated tuple with the new score (note the += operator)
- line 12-13: if this is a new student (unknown to the dictionary), create a new entry - its value is a one-element tuple containing the entered score;
- line 15: iterate through the sorted students' names;
- line 16-17: initialize the data needed to evaluate the average (sum and counter)
- line 18-20: we iterate through the tuple, taking all the subsequent scores and updating the sum, together with the counter;
- line 21: evaluate and print the student's name and average score.

````python
schoolClass = {}

while True: #Looping, Program akan terus mencetak pertanyaan yang sama

    name = input("Enter the student's name (or type exit to stop): ")
    if name == 'exit':
        break
        #Looping berhenti apabila user menginputkan "exit"

    score = int(input("Enter the student's score (0-10): "))

    if name in schoolClass: #Checking nama murid di function
        schoolClass[name] += (score,)
    else:
        schoolClass[name] = (score,)

print(schoolClass)

for name in sorted(schoolClass.keys()): #Mencetak nama murid urut abjad
    sum = 0 #Jumlah nilai
    counter = 0 #Banyak nilai yang masuk/diinput
    #Kalkulator menghitung jumlah nilai dan inputan yang masuk
    for score in schoolClass[name]:
        sum += score
        counter += 1
    print(name, ":", sum / counter) #Perhitungan rat-rata nilai
````

````python
def tf(sentence):
  #Fungsi menghitung frekuensi kemunculan setiap kata dalam sebuah kalimat
    d = {}
    for word in sentence.split():#Memotong kalimat menjadi perkata dan di cek satu persatu
        if word not in d:
            d[word] = 1
        #Jika ada kata yang belum ada di dict, maka tambahkan sebagai key baru bernilai 1
        else:
            d[word] += 1
            #Jika sudah ada, maka nilainya +1
    return d

kalimat = "Ibu pergi ke pasar dengan Bapak dan Ibu Ibu lain" #Testing
stopw.append('dan')
stopw.append('lain')
#Kalimat yang di blacklist (tidak ditampilkan)
kalimat_stop = removeStop(kalimat.lower())
#Melakukan penghapusan pada kata yang terdapat pada stopw, dan mengubah semua kata ke lower
tf_kalimat = tf(kalimat_stop)

print(tf_kalimat)
````
