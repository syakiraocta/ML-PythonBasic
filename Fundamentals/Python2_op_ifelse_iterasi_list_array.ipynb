## Struktur Pemilihan

Berbeda dengan bahasa pemrograman lainnya, Python hanya memiliki satu statement untuk melakukan proses pemilihan, yaitu dengan menggunakan if. Untuk mempermudah pembahasan tentang statement if, pembahasan akan dibagi menjadi tiga kelompok :

 - Statement if untuk satu kasus
 - Statement if untuk dua kasus
 - Statement if untuk tiga kasus

### 1. Statement if untuk satu kasus

Bentuk umum penggunaan perintah if untuk satu kasus dalam Python adalah :

```python
if kondisi:
    statement1
    statement2
    ...
    
```
Pada bentuk diatas, statement1, statement2 dan seterusnya hanya akan dieksekusi jika kondisi yang didefinisikan dalam perintah If terpenuhi (bernilai True). Jika kondisi bernilai False, statement tersebut akan diabaikan oleh program.

Perhatikan contoh berikut :

````python
a = 5
b = 6

if a==b:
    print("oke")
    print("berhasil")
    print("Gagal")
````

Nilai dikedua variabel berbeda a= 5 dam b =6 maka hasilnya/outpunya gagal, karena tidak memenuhi kondis(nilainya tidak sama)

### 2. Statement if untuk dua kasus

Untuk pemilihan yang mengandung dua kasus, bentuk umum yang digunakan adalah :

```python
if kondisi :
    statement1
    statement2
    ...
else
    statement_alternatif1
    statement_alternatif2
    ...
```

Pada bentuk ini, jika kondisi bernilai True makan statement selanjutnya yang akan dieksekusi oleh program adalah statement1, stetement 2 dan seterusnya. Tapi jika kondisi bernilai False, maka program akan mengeksekusi statement yang berada pada bagian else, yaitu statement_alternatif1, statement_alternatif2 dan seterusnya.

````python
a = 5
b = 6

if a==b:
    print("oke")
    print("berhasil")
else:
    print("Gagal")
````

Dalam statment if untuk dua kasus, jika kondisi bernilai true maka program akan menghasilkan nilai yang berada di if. Sedangkan jika tidak memenuhi nilai yang ada di if maka output yang dihasilkan adalah nilai yang ada di else, bedanya dengan Statment if untuk satu kasus adalah untuk yang dua kasus nilai "gagal" masuk kedalam else atau rencana cadangan. Statment ini mengharuskan program memilih jawaban dari salah satu inputan yang ada, sedangkan pada statment satu bisa saja ouput tidak ada karena tidak ada inputan yang sesuai.

Struktur pemilihan jenis ini merupakan bentuk yang paling kompleks jika dibandingkan dengan kedua jenis yang sudah di bahas diatas. Dalam struktur ini, ada beberapa kondisi yang harus diperiksa oleh program. Bentuk umum penggunaan perintah if untuk tiga kasus atau lebih adalah :

```python
if kondisi1:
    statement1a
    statement1b
    ...
elif kondisi2:
    statement2a
    statement2b
    ...
else:
    statement_alternatif1
    statement_alternatif2
    ...
```
Perhatikan contoh dibawah ini :

Dalam statment if untuk tiga kasus, terdapat tiga program yang menjadi pilihan yaitu  
- If  => Kondisi yang utama (didahulukan)  
- Elseif  => Kondisi dimana kondisi utama tidak memenuhi syarat  
- Else => Kondisi terakhir yang menandakan bahwa program tidak memenuhi kondisi utama dan kondisi cadangan (Elseif)  

Statmen ini di gunakan apabila kita membutuhkan data yang perlu banyak pilihan opsi

- otherwise, if the year number isn't divisible by 100, it's a leap year;
- otherwise, if the year number isn't divisible by 400, it's a common year;

#### 3.1 Statement break

Statement break digunakan untuk menghentikan secara paksa proses perulangan, meskipun kondisi perulangan sebenarnya masih bernilai True. Perhatikan contoh-contoh berikut ini :

````python
for i in range (11):
    print(i, end=" ") #program agar tidak dicetak kebawah tapi kesamping
    if i==7: #program tambahan sebagi syarat berhenti diantrian tsb
        print(" masih dalam for")
        break #Fungsi menghentikan antrean looping
````

````python
# break - example

print("The break instruction:")
for i in range(1, 6):
    if i == 3:
        break
    print("Inside the loop.", i)
print("Outside the loop.")
````

Kondisi dimana progrram memiliki range (1,6) tetapi ada syarat harus berhenti di i==3, maka apabila program melalkukan looping dengan nilai dibawah 3 hasil outputnya akan "inside the loop".
````python
largestNumber = -99999999 #Patokan angka terbesar dalma program
counter = 0 #Jumlah inputan yang dilakukan

while True: #Looping inputan menu angka
    number = int(input("Enter a number or type -1 to end program: "))
    if number == -1: #Syarat antrean looping dianggap break
        break
    counter += 1 #Section menu inputan
    if number > largestNumber:
    #Pengecekan antara inputan dan largestNumber, yang lebih besar
        largestNumber = number
        #Jika lebih besar maka angka terbesar diganti dengan angka baru

if counter != 0:
    print("The largest number is", largestNumber)
else:
    print("You haven't entered any number.")
    #Kondisi dimana langsung meninput angka -1 (tidak ada pembanding)
````

````python
# continue - example

print("\nThe continue instruction:")
for i in range(1, 6):
    if i == 3:
        continue
    print("Inside the loop.", i)
print("Outside the loop.")
````

Dalam kondisi ini memiliki syarat apabila i == 3 maka program continue (lompat) ke antrean beriktnya dan karena batas akhirnya 6, mak aprogram dengan antrean di atas 5 di anggap outside the loop.

````python
largestNumber = -99999999
counter = 0

number = int(input("Enter a number or type -1 to end program: "))

while number != -1:
    if number == -1:
        continue
    counter += 1

    if number > largestNumber:
        largestNumber = number
    number = int(input("Enter a number or type -1 to end program: "))

if counter:
    print("The largest number is", largestNumber)
else:
    print("You haven't entered any number.")
````

#### Bubble Short


````python
myList = [8, 10, 6, 2, 4] # list to sort
swapped = True # it's a little fake - we need it to enter the while loop

while swapped:
    swapped = False # no swaps so far
    for i in range(len(myList) - 1):#Program checking dari urutan akhir/belakang
        if myList[i] > myList[i + 1]:
          #Checking mana yang lebih besar antara antrean 1 dengan yang sebelahnya
            swapped = True # swap occured!
            myList[i], myList[i + 1] = myList[i + 1], myList[i]

print(myList)
````

List diatas menggambarkan list papan catur yang masih kosong sebagaimana gambar ilustrasi dibawah :



Mari kita isi list sersebut dengan sesuatu yang kita inginkan. Perhatikan kode berikut :

````python
EMPTY = "-"
LIMALIMA = "LIMA"
ROOK = "ROOK"
KNIGHT = "KNIGHT"
PAWN = "PAWN"
board = []

for i in range(8):
    row = [EMPTY for i in range(8)]
    board.append(row)

#isi dengan ROOK
board[0][0] = ROOK
board[5][5] = LIMALIMA
board[0][7] = ROOK
board[7][0] = ROOK
board[7][7] = ROOK

#isi dengan KNIGHT
board[4][2] = KNIGHT

#isi dengan PAWN
board[3][4] = PAWN


print(board)
print("\n")
````

Maka total data yang akan direkam adalah 24 jam x 31 hari = 744 data.

Pertama-tama, kita harus memutuskan tipe data yang akan tepat untuk aplikasi ini. Selanjutnya, kita akan menentukan bahwa baris akan merepresentasikan jam (sehingga ada 24 baris) dan setiap baris akan direpresentasikan untuk satu hari (sehingga kita membutuhkan lagi 31 baris).

Menampilkan list cuaca selama 4 hari, dan setiap sehari memiliki 2 kolom jam

````python
temps = [[0.0 for h in range(24)] for d in range(31)]

# 1. the monthly average noon temperature
sum = 0.0

for day in temps:
    sum += day[11]

average = sum / 31

print("Average temperature at noon:", average)

# 2. the highest temperature during the whole month
highest = -100.0

for day in temps:
    for temp in day:
        if temp > highest:
            highest = temp

print("The highest temperature was:", highest)

# 3.count the days when the temperature at noon was at least 20 ℃
hotDays = 0

for day in temps:
    if day[11] > 20.0:
        hotDays += 1

print(hotDays, "days were hot.")
````

#### 3. Array tiga dimensi

Python tidak memiliki batasan dalam list-in-list. Ini merupakan contoh untuk array tiga dimensi.

Bayangkan sebuah hotel yang besar terdiri dari tiga bangunan, 15 lantai untuk tiap  bangunannya. Pada setiap lantai terdapat 20 kamar.

Langkah pertama adalah tentukan tipe data dari array yang akan dibuat. Dalam kasus ini nilai Boolean yang paling sesuai, nilai True jika kamar telah di pesan dan nilai False jika kamar kosong. Langkah kedua adalah dengan menganalisa kondisi yang diberikan.

Kode dibawah akan menghasilkan array :

````python
rooms = [[[False for r in range(2)] for f in range(3)] for t in range(4)]
print (rooms)
````

Coba berikan input untuk kondisi bibawah ini :
1. Pesan satu kamar pada gedung 2, lantai 10, kamar 14
2. Kosongkan satu kamar pada gedung 1, lantai 5, kamar 2
3. Periksa apakah ada kamar kosong di lantai 15

````python
# --- STEP 1: MEMBUAT GEDUNG (LIST 3D) ---
# Struktur: [Gedung][Lantai][Kamar]
# 3 Gedung, 15 Lantai, 20 Kamar per lantai. Status awal False (Kosong).
rooms = [[[False for r in range(20)] for f in range(15)] for t in range(3)]

# --- STEP 2: PROSES TRANSAKSI KAMAR ---

# 1. Pesan satu kamar pada gedung 2, lantai 10, kamar 14
# Index: Gedung[1], Lantai[9], Kamar[13]
rooms[1][9][13] = True
print("Kamar di Gedung 2, Lantai 10, Kamar 14 telah dipesan.")

# 2. Kosongkan satu kamar pada gedung 1, lantai 5, kamar 2
# Index: Gedung[0], Lantai[4], Kamar[1]
rooms[0][4][1] = False
print("Kamar di Gedung 1, Lantai 5, Kamar 2 telah dikosongkan.")

print("-" * 30)

# --- STEP 3: CEK KETERSEDIAAN (VACANCY) ---

# 3. Periksa apakah ada kamar kosong di lantai 15 (Gedung ke-3 / Index 2)
vacancy = 0
for roomNumber in range(20):
    # 'if not' artinya mengecek yang nilainya False (Kosong)
    if not rooms[2][14][roomNumber]:
        vacancy += 1

# Cetak hasil pengecekan
print(f"Hasil Pengecekan di Gedung 3, Lantai 15:")
print(f"Jumlah kamar yang masih kosong: {vacancy} kamar.")

# --- STEP 4: CEK STATUS SPESIFIK ---
# Coba kita cek kamar yang tadi kita pesan di Step 2 nomor 1
if rooms[1][9][13] == True:
    print("\nStatus Kamar [1][9][13]: TERISI (Booked)")
else:
    print("\nStatus Kamar [1][9][13]: KOSONG (Available)")
````
