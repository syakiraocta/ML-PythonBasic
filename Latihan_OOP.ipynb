## Parent Class

```python

class Person:
    def __init__(self, name, address):
        # Menggunakan __ agar atribut bersifat private (Encapsulation)
        self.__name = name
        self.__address = address

    # Getter untuk mengambil nama
    def getName(self):
        return self.__name

    # Getter untuk mengambil alamat
    def getAddress(self):
        return self.__address

    # Setter untuk mengubah/update nama
    def setName(self, name):
        self.__name = name

    # Method spesial untuk menampilkan identitas dalam bentuk teks
    def __str__(self):
        return f"{self.__name}, Alamat: {self.__address}"
```

## Class Student (Inheritance dari Person)  
Class ini mewarisi Nama dan Alamat, lalu menambah fitur Nilai dan Mata Pelajaran.

```python
# --- CLASS STUDENT ---
class Student(Person):
    def __init__(self, name, address):
        super().__init__(name, address) # Manggil bapaknya (Person)
        self.__numCourses = 0
        self.__courses = []
        self.__grades = []

    def addCourseGrade(self, course, grade):
        self.__courses.append(course)
        self.__grades.append(grade)
        self.__numCourses += 1

    def printGrades(self):
        for i in range(len(self.__courses)):
            print(self.__courses[i] + " : " + str(self.__grades[i]))

    def getAverageGrade(self):
        total = 0 # Ganti 'sum' biar gak bentrok sama fungsi bawaan
        for n in self.__grades:
            total += n
        return total / len(self.__grades)

    def __str__(self):
        return "Student: " + super().getName() + " " + super().getAddress()
```

## Class Teacher (Inheritance dari Person)  
Class ini mewarisi Nama dan Alamat, tapi fiturnya adalah menambah/menghapus mata pelajaran yang diajarkan.

```python
# --- CLASS TEACHER ---
class Teacher(Person):
    def __init__(self, name, address):
        super().__init__(name, address)
        self.__numCourses = 0
        self.__courses = []

    def addCourse(self, course):
        if course not in self.__courses:
            self.__courses.append(course)
            self.__numCourses += 1 # Tambah hitungan kursus
            return True
        else:
            return False

    def removeCourse(self, course):
        if course in self.__courses:
            i = self.__courses.index(course)
            del self.__courses[i]
            self.__numCourses -= 1 # Kurangi hitungan kursus
            return True
        else:
            return False

    def printCourse(self):
        print(self.__courses)

    def __str__(self):
        return "Teacher: " + super().getName() + " " + super().getAddress()
```

# Testing Output Program

```python
# Test Guru
t1 = Teacher("Pak Budi", "Jakarta")
print(t1) # Output: Teacher: Pak Budi Jakarta

t1.addCourse("Matematika")
t1.addCourse("Fisika")
t1.printCourse() # Output: ['Matematika', 'Fisika']

# Test Siswa
s1 = Student("Syakira", "Pekalongan")
print(s1) # Output: Student: Syakira Pekalongan
s1.addCourseGrade("Pemrograman", 95)
s1.addCourseGrade("Basis Data", 90)
s1.printGrades()
print("Rata-rata:", s1.getAverageGrade())
```
